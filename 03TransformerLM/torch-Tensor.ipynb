```python
import torch
```

### `torch.stack` 按维度堆叠

```python
a = torch.ones(1, 3, 4)
b = torch.ones(1, 3, 4) * 2
```

```python
x = torch.stack((a, b), dim=0)
x, x.shape
```

```python
x = torch.stack((a, b), dim=1)
x, x.shape
```

```python
x = torch.stack((a, b), dim=-1)
x, x.shape
```

### `torch.flatten` 展平

```python
a = torch.randn(2, 3, 4, 5)
a.shape
```

```python
x = a.flatten(1, 2)
x.shape
```

```python
# pytorch 中 - 表示从后向前数 
x = a.flatten(-2) # 合并后两维
x.shape
```
